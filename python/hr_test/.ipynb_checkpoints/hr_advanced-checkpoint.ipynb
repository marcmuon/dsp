```python
import csv
import re

def count_degrees(csv_file_name):
    '''read in csv and return dict with frequency of degrees after data clean-up'''
    
    file = open(csv_file_name, 'r')
    reader = csv.reader(file)
    
    degrees = []
    for row in reader:
        degrees.append(row[1].strip())
    degrees.pop(0)
    
    degree_unpack = []
    p1 = re.compile('\s')
    for degree in degrees:
        degree_unpack.extend(p1.split(degree))
    
    degree_final = []
    p2 = re.compile('\.')
    for degree in degree_unpack:
        degree_final.append(p2.sub('', degree))
    
    degree_dict = {}
    for degree in degree_final:
        degree_dict[degree] = degree_dict.get(degree,0) + 1
        
    return degree_dict
```

```python
import csv
import re

def count_titles(csv_file_name):
    '''read in csv and return dict with frequency of titles after data clean-up'''
    
    file = open(csv_file_name, 'r')
    reader = csv.reader(file)
    
    titles = []
    for row in reader:
        titles.append(row[2])
    titles.pop(0)
    
    titles_clean = []
    p = re.compile(r'\bis\b')
    for title in titles:
        titles_clean.append(p.sub('of',title))
    
    title_dict = {}
    for title in titles_clean:
        title_dict[title] = title_dict.get(title,0) + 1
        
    return title_dict
```

```python
import csv
import re

def emails(csv_file_name):
    
    file = open(csv_file_name,'r')
    reader = csv.reader(file)
    
    p = re.compile(r'\w*@(\w|\.)*')
    email_list = []
    for row in reader:
        match = p.search(str(row))
        if match:
            email_list.append(match.group())
    return email_list
```

```python
e = ['rxiao@mail.med.upenn.edu',
 'sxie@mail.med.upenn.edu',
 'dxie@upenn.edu',
 'weiyang@mail.med.upenn.edu']

import re
def unique_domains(emails):
    '''Given a list of emails return the unique domains'''
    p = re.compile(r'\w+@')
    domain_only = []
    for address in emails:
        domain_only.append(p.sub('',address))
    return list(set(domain_only))
unique_domains(e)
```

```python
# read faculty.csv

# Need to aggregate SAME last names

import csv
import re

def get_dict():
    file = open('faculty.csv','r')
    reader = csv.reader(file)
    d = {}
    p = re.compile(r'\S+$')
    for row in reader:
        match = p.search(row[0])
        if match:
            row[0] = match.group()
        if row[0] not in d:
            d[row[0]] = [[row[1],row[2],row[3]]]
        else:
            d[row[0]].extend([[row[1],row[2],row[3]]])
    del d['name']
    return d
get_dict()
```
